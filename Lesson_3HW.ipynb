```python
import requests
from bs4 import BeautifulSoup as bs
from selenium  import webdriver
```

```python
ref="https://rskrf.ru/ratings/produkty-pitaniya/"
```

```python
my_products={"сайт":ref}
```

```python
products_ref=[]
```

```python
def search_for_products(products_ref,reference,height=0):
    if height==2:
        products_ref.append(reference)
        return
    driver = webdriver.Chrome()
    driver.get(reference)
    soup=bs(driver.page_source,'lxml')
    categories=soup.find('div',attrs={'class':'categories'})
    for category in categories.findChildren("a"):
        search_for_products(products_ref,reference + category["href"].split("/")[-2]+"/",height+1)
```

```python
search_for_products(products_ref,ref)
```

```python
def fill_characteristic(ref,name):
    res=requests.get(ref, headers={
            'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'
        })
    soup=bs(res.content,'html.parser')
    parameters=soup.findChildren("div",class_='rating-item')
    my_products[name]=list()
    for p in parameters:
        my_products[name].append(p.text.replace('\n',' '))
```

```python
def fill_map_of_products(products):
    if products is None:
        return
    for pr in products.findChildren("div",class_="outer"):
        fill_characteristic("https://rskrf.ru"+pr.find("a")["href"],pr.find("a").text)
```

```python
for product in products_ref:
    driver = webdriver.Chrome()
    driver.get(product)
    soup=bs(driver.page_source,'lxml')
    categories=soup.find("div", attrs={"class":'goods_list'})
    fill_map_of_products(categories)
```
